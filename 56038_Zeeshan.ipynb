```python
import numpy as np
#numpy array slicing
d1 = np.array([6,8,1,2,54,5,3]) #1-D Array
print("Elements of array d1 are ", d1)
print("Elements slicing 'd1[1:5]'", d1[1:5]) # 1-D Array slicing "The result includes the start index, excludes the end index"
print("Elements slicing d1[:3] ", d1[:3]) #Slice elements form start index[0] to index[3], index [3] excluded
print("Elements slicing d1[3:] ", d1[3:]) #Slice elements form index[3] to last index, last Index excluded
print("Elements slicing 'd1[-3:-1]'", d1[-3:-1]) #slice elements (negative slicing) form index[-3] to index[-1], index[-1] excluded
d2 = np.array([[5.9,3,4],[21,65,8.09],[56.08,1,2]]) #2-D Array
print("Elements of array d2 are ", d2)
print("Elements slicing 'd2[0,1:3]'", d2[0,1:3]) # 2-D Array slicing: Slice elements form array[0], Index[1] to Index[3], index [3] excluded


# Array slicing with step [index:index: step]
# print("Elements of array d1 are ",d1)
print("Array d1 slicing with step: d1[1:6:2] ",d1[1:6:2])
```

```python
import numpy as np

# Create a 3x3 matrix with values ranging from 2 to 10
matrix = np.arange(2, 11).reshape(3, 3)

# Convert array to float type
matrix = matrix.astype(float)

# Flatten the array
flattened = matrix.flatten()

# Find maximum and minimum values
max_value = np.max(flattened)
min_value = np.min(flattened)

print("Matrix:\n", matrix)
print("Flattened array:", flattened)
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```
